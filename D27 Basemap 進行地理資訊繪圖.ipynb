```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 創建基本地圖。

plt.figure(figsize=(5,5))
m=Basemap(projection="ortho", lat_0=0, lon_0=0)
m.bluemarble()

plt.show()
```

```python
m = Basemap(width=12000000,height=9000000,projection='lcc',resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107)
m.bluemarble()
plt.show()
```

```python
m = Basemap(projection='mill',llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180) # 成圖設定resolution 

m.drawcoastlines(linewidth=2)
m.drawcountries()
m.drawstates(color='r')
m.drawcounties(color='darkred')

plt.title('Basemap Tutorial')
plt.show()
```
