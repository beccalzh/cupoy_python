```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. 畫出 cos 圖檔，並儲存
x = np.arange(0,540)
y = np.cos(x * np.pi / 180.0)

plt.plot(x, y, color="#ADADAD")
plt.xlim(-30,600)
plt.ylim(-1.5,1.5)

plt.xlabel("x-axis") 
plt.ylabel("y-axis") 
plt.title("Line Chart") 

plt.savefig("filename.png",dpi=300,format="png")
```

```python
#2. 給定散點圖顏色
c = np.random.normal(0,1,500)
d = np.random.normal(0,1,500)
plt.scatter(c, d, c='#00CACA')
plt.title("Scatter plot")
```
