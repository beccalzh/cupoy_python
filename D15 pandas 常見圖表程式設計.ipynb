```python
import pandas as pd
import numpy as np
df = pd.read_csv('boston.csv')
df
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
dff=[]
for i in list(df.columns):
    if df[i].mean()>100:
        dff.append(i)
df1=df.loc[:,dff]
df1.boxplot() #結果顯示，TAX及B的中位數介於300~400
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
df.plot.scatter(x='NOX',y='DIS') #圖片說明兩個變數呈現負相關，即NOX增加DIS會減少
```
